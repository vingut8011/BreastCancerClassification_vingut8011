```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df.drop(columns=["Diagnosis"])
y = df["Diagnosis"]

encoder = LabelEncoder()
y = encoder.fit_transform(y)

scaler = StandardScaler()

X[["concavity1", "concavity3"]] = scaler.fit_transform(
    X[["concavity1", "concavity3"]]
)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.30,
    random_state=42
)

print("Training Data Shape:", X_train.shape)
print("Testing Data Shape:", X_test.shape)
```

# Step 5: Implement SVM

Date: 05/24/2026

Student ID: vingut8011

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

svm_model = SVC()

svm_model.fit(X_train, y_train)

svm_predictions = svm_model.predict(X_test)

print("SVM Accuracy Score:")
print(accuracy_score(y_test, svm_predictions))

print("\nClassification Report:")
print(classification_report(y_test, svm_predictions))
```

# Step 6: Apply GridSearchCV for SVM Hyperparameter Tuning

Date: 05/24/2026

Student ID: vingut8011

```python
from sklearn.model_selection import GridSearchCV

# Parameter grid
svm_param_grid = {
    "C": [0.1, 1, 10, 100],
    "kernel": ["linear", "rbf", "poly"]
}

# GridSearchCV
svm_grid = GridSearchCV(
    SVC(),
    svm_param_grid,
    cv=5
)

# Fit model
svm_grid.fit(X_train, y_train)

# Best parameters
print("Best SVM Parameters:")
print(svm_grid.best_params_)

# Best model
best_svm = svm_grid.best_estimator_

# Predictions
tuned_svm_predictions = best_svm.predict(X_test)

# Accuracy
print("\nTuned SVM Accuracy Score:")
print(accuracy_score(y_test, tuned_svm_predictions))

# Classification report
print("\nTuned SVM Classification Report:")
print(classification_report(y_test, tuned_svm_predictions))
```

# Step 7: Implement XGBoost

Date: 05/24/2026

Student ID: vingut8011

```python
from xgboost import XGBClassifier

xgb_model = XGBClassifier(
    eval_metric="logloss",
    random_state=42
)

xgb_model.fit(X_train, y_train)

xgb_predictions = xgb_model.predict(X_test)

print("XGBoost Accuracy Score:")
print(accuracy_score(y_test, xgb_predictions))

print("\nClassification Report:")
print(classification_report(y_test, xgb_predictions))
```

# Step 8: Apply GridSearchCV for XGBoost Hyperparameter Tuning

Date: 05/24/2026

Student ID: vingut8011

```python
from sklearn.model_selection import GridSearchCV

# Parameter grid
xgb_param_grid = {
    "n_estimators": [50, 100],
    "learning_rate": [0.1, 0.2],
    "max_depth": [3, 4]
}

# GridSearchCV
xgb_grid = GridSearchCV(
    XGBClassifier(
        eval_metric="logloss",
        random_state=42
    ),
    xgb_param_grid,
    cv=5
)

# Fit model
xgb_grid.fit(X_train, y_train)

# Best parameters
print("Best XGBoost Parameters:")
print(xgb_grid.best_params_)

# Best model
best_xgb = xgb_grid.best_estimator_

# Predictions
tuned_xgb_predictions = best_xgb.predict(X_test)

# Accuracy
print("\nTuned XGBoost Accuracy Score:")
print(accuracy_score(y_test, tuned_xgb_predictions))

# Classification report
print("\nTuned XGBoost Classification Report:")
print(classification_report(y_test, tuned_xgb_predictions))
```
